Canales y colores Reales

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize']=[10,5]
img=cv2.imread('1.jpg',cv2.IMREAD_COLOR)
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
r=img[:,:,2]
g=img[:,:,1]
b=img[:,:,0]
r_imgen=np.zeros_like(img_rgb)
r_imgen[:,:,0]=r
g_imgen=np.zeros_like(img_rgb)
g_imgen[:,:,1]=g
b_imgen=np.zeros_like(img_rgb)
b_imgen[:,:,2]=b

plt.subplot(141)
plt.title('Original')
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(142)
plt.title('CAnal rojo')
plt.imshow(r_imgen)
plt.axis('off')

plt.subplot(143)
plt.title('CAnal Verde')
plt.imshow(g_imgen)
plt.axis('off')

plt.subplot(144)
plt.title('CAnal Azul')
plt.imshow(b_imgen)
plt.axis('off')

plt.show()
```

Colores Falsos

```python
import cv2
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize']=[10,5]
img=cv2.imread('1.jpg',cv2.IMREAD_COLOR)
b,g,r=cv2.split(img)

plt.subplot(141)
plt.title('Original')
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(142)
plt.title('CANAL ROJO')
plt.imshow(r,cmap='Reds')
plt.axis('off')

plt.subplot(143)
plt.title('CANAL VERDE')
plt.imshow(g,cmap='Greens')
plt.axis('off')

plt.subplot(144)
plt.title('CANAL AZUL')
plt.imshow(b,cmap='Blues')
plt.axis('off')
plt.show()
```

ESCALA DE GRISES

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np


image = cv2.imread('1.png', cv2.IMREAD_GRAYSCALE)

img_inicial = image.dtype
print('Profundidad de color original:', img_inicial)

img_16bit = np.int16(image)
print('Profundidad de color nueva:', img_16bit.dtype)


plt.figure(figsize=(12, 4))


plt.subplot(1, 2, 1)
plt.imshow(image) 
plt.axis('off')
plt.title('Imagen original')

plt.subplot(1, 2, 2)
plt.imshow(img_16bit)
plt.axis('off')
plt.title('Imagen 16 bits')

plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

img_bgr = cv2.imread('1.png')  
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)  
hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)  

fig, axs = plt.subplots(1, 4, figsize=(25, 10))

axs[0].imshow(img_rgb)
axs[0].set_title('imagen')



fig_img1 = axs[1].imshow(hsv[...,0],cmap='gray', vmin=0, vmax=179)
fig.colorbar(fig_img1, ax=axs[1], shrink=0.3)
axs[1].set_title('canal H')

fig_img2 = axs[2].imshow(hsv[...,1],cmap='gray', vmin=0, vmax=255)
fig.colorbar(fig_img1, ax=axs[2], shrink=0.3)
axs[2].set_title('canal S')

fig_img3 = axs[3].imshow(hsv[...,0],cmap='gray', vmin=0, vmax=255)
fig.colorbar(fig_img1, ax=axs[3], shrink=0.3)
axs[3].set_title('canal V')
```
