```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# Definicin de colores
colores = [(0, 'purple'), (0.4, 'blue'), (0.7, 'green'), (1, 'yellow')] 
cmap = LinearSegmentedColormap.from_list('degradado', colores)  

imagen_degra = cmap(np.linspace(0, 1, 650).reshape(-1, 1))

plt.figure(figsize=(12, 5))

plt.imshow(imagen_degra, aspect='auto')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

tamao = 100
filas, conlumnas = 8, 8

tablero = np.zeros((filas*tamao, conlumnas*tamao), dtype=np.uint8)
for i in range(filas):
    for j in range(conlumnas):
        if ( i + j ) % 2 == 0:
            tablero [i*tamao:(i+1)*tamao, j*tamao:(j+1)*tamao] = 255
 

plt.imshow(tablero, cmap='gray')
plt.title('Tablero de Ajedrez')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('imagen1.jpg', 0)
degradado = np.linspace(1, 0, imagen.shape[0])

mascara = np.tile(degradado, (imagen.shape[1], 1)).T #Imagen transcpuesta
imagen_normalizada = cv2.normalize(imagen, None, 0, 1, cv2.NORM_MINMAX)

imagen_degradada = imagen_normalizada * mascara

plt.figure(figsize=(10,5))

plt.subplot(121)
plt.imshow(imagen, cmap='gray')
plt.title('Imagen Original')

plt.subplot(122)
plt.imshow(imagen_degradada, cmap='gray')
plt.title('Imagen con Degradado Lineal (De abajo a arriba)')

plt.show()
```
